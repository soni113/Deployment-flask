```python
face_cascade=cv2.CascadeClassifier('/content/data.xml')
```

```python
faces=face_cascade.detectMultiScale(gray,1.1,4)
for (x,y,w,h) in faces:
  cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,0),2)
  cv2_imshow(image)
  cv2.waitkey(0)
  cv2.destroyAllWindows()
```
