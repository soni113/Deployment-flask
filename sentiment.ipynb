```python
word_index=dict(zip(word_index.Words,word_index.Indexes))
```

```python
word_index["<PAD>"]=0
word_index["<START"]=1
word_index["<UNK>"]=2
word_index["<UNUSED>"]=3
```

```python
def review_encoder(text):
  arr=[word_index[word] for word in text]
  return arr
```

```python
train_data,train_labels=imdb_review['Reviews'],imdb_review['Sentiment']
test_data, test_labels=test_review['Reviews'],test_review['Sentiment']
```

```python
train_data=train_data.apply(lambda review:review.split())
test_data=test_data.apply(lambda review:review.split())
```
